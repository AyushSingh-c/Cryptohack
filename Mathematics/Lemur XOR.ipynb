```python
from PIL import Image, ImageChops
im1 = Image.open('Lemur_XOR/lemur.png')   #path for lemur image
im2 = Image.open('Lemur_XOR/flag.png')    #path for flag image 

im3 = ImageChops.add(ImageChops.subtract(im2, im1), ImageChops.subtract(im1, im2))
im3.show()
im3.save("Lemur_XOR/im3.png")      #path for final decoded image


import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('Lemur_XOR/im3.png')
imgplot = plt.imshow(img)
plt.show()
```
